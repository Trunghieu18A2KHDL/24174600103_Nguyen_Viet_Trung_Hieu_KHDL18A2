```python
from xml.dom import minidom

# Đọc và phân tích file XML
doc = minidom.parse("sample.xml")

# Lấy tất cả các phần tử (node) có tên 'name'
elements = doc.getElementsByTagName("name")

# In ra tên của từng phần tử
print("Danh sách các phần tử <name>:")
for elem in elements:
    # Lấy nội dung text trong thẻ <name>
    print(elem.firstChild.data)
```
