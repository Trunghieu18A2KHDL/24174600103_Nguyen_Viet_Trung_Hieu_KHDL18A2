```python
class Stack:
    def __init__(self, max_size):
        self.max_size = max_size
        self.stack = []
        self.count = 0  # đếm số phần tử hiện có

    def push(self, value):
        if not self.isFull():
            self.stack.append(value)
            self.count += 1
        else:
            print("Ngăn xếp đã đầy!")

    def pop(self):
        if not self.isEmpty():
            self.count -= 1
            return self.stack.pop()
        else:
            print("Ngăn xếp rỗng!")
            return None

    def isEmpty(self):
        return self.count == 0

    def isFull(self):
        return self.count == self.max_size

    # Hàm trả về số phần tử
    def getCount(self):
        return self.count


# --- Chương trình chính ---
if __name__ == "__main__":
    st = Stack(5)
    st.push(1)
    st.push(2)
    st.push(3)
    print("Số phần tử hiện có trong stack:", st.getCount())

    st.pop()
    print("Sau khi pop, số phần tử:", st.getCount())
```
