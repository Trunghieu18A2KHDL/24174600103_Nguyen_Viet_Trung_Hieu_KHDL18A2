```python
class Stack:
    def __init__(self, max_size):
        self.max_size = max_size
        self.stack = []

    # Thêm phần tử vào đỉnh stack
    def push(self, value):
        if not self.isFull():
            self.stack.append(value)
        else:
            print("Ngăn xếp đã đầy, không thể thêm phần tử!")

    # Lấy phần tử ra khỏi đỉnh stack
    def pop(self):
        if not self.isEmpty():
            return self.stack.pop()
        else:
            print("Ngăn xếp rỗng, không thể lấy phần tử!")
            return None

    # Kiểm tra rỗng
    def isEmpty(self):
        return len(self.stack) == 0

    # Kiểm tra đầy
    def isFull(self):
        return len(self.stack) == self.max_size


# --- Chương trình chính ---
if __name__ == "__main__":
    st = Stack(3)
    st.push(10)
    st.push(20)
    st.push(30)
    st.push(40)  # quá sức chứa

    print("Phần tử lấy ra:", st.pop())
    print("Stack còn lại:", st.stack)
```
