```python
class PhanSo:
    # Hàm khởi tạo
    def __init__(self, tu=0, mau=1):
        self.tu = tu
        self.mau = mau

    # Kiểm tra tính hợp lệ (mẫu số ≠ 0)
    def hop_le(self):
        return self.mau != 0

    # Nhập phân số
    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        while self.mau == 0:
            print("Mẫu số phải khác 0! Nhập lại.")
            self.mau = int(input("Nhập mẫu số: "))

    # In phân số
    def xuat(self):
        if self.hop_le():
            print(f"Phân số: {self.tu}/{self.mau}")
        else:
            print("Phân số không hợp lệ.")


# --- Chương trình chính ---
if __name__ == "__main__":
    ps = PhanSo()
    ps.nhap()
    ps.xuat()
```
